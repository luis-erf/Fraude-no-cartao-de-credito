```python
#- Imports e configurações
import warnings
import pandas as pd

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

#Análise de valores faltantes


```python
# 1. Contagem de missing values
print("\n1️⃣ MISSING VALUES POR COLUNA:")
missing_data = pd.DataFrame({
    'Coluna': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percent': (df.isnull().sum() / len(df)) * 100
})
missing_data = missing_data[missing_data['Missing_Count'] > 0].sort_values('Missing_Percent', ascending=False)
print(missing_data)
```

Assim é possível perceber que esse dataframe não possui valores faltantes, logo não é necessário fazer nenhum tratamento para esse tipo

#Análise de duplicatas

```python
print("\n6️⃣ VALORES ÚNICOS POR COLUNA:")
for col in df.columns:
    unique_count = df[col].nunique()
    print(f"{col}: {unique_count} valores únicos")
```

Ao analisar os valores únicos percebe-se que id só possui valores únicos que servem como identificação das transações, logo não é importante para o modelo, o que nos permite dropar essa feature

```python
df.drop(columns=['id'], inplace=True)
```

Como todas as features do tipo V apresentaram a mesma quantidade de valores únicos e eles não são a mesma quantidade de observações. Isso pode ser algo importante para ser análisado na Análise Exploratória de Dados

```python
df_duplicates = df[df.duplicated(keep=False)]
df_duplicates.head()
```

```python
df_duplicates.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
import joblib
joblib.dump(df, 'train_prep.csv')
```

#Conclusão

Não foi necessário o imputamente de valores faltantes, pois o dataset não possui valores faltantes e foram removidos os valores duplicados
